```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
```

```python
spaceship = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/spaceship_titanic.csv")
spaceship.head()
```

Now perform the same as before:
- Feature Scaling
- Feature Selection


```python
spaceship = spaceship.dropna()
```

```python
spaceship['Cabin'] = spaceship['Cabin'].apply(lambda x: x[0] if pd.notna(x) else None)
```

```python
spaceship = spaceship.drop(columns=['PassengerId', 'Name'])
```

```python
non_numeric_spaceship = spaceship.select_dtypes(exclude=['number'])

column_summary_spaceship = pd.DataFrame({
    'Column Name': non_numeric_spaceship.columns,
    'Data Type': non_numeric_spaceship.dtypes.values,
    'Percentage Null': non_numeric_spaceship.isnull().mean().values * 100,
    'Unique Values': non_numeric_spaceship.nunique().values
})

print(column_summary_spaceship)
```

```python
non_numeric_cols = spaceship.select_dtypes(exclude=['number'])
df_dummies = pd.get_dummies(non_numeric_cols)
spaceship_with_dummies = pd.concat([spaceship.drop(non_numeric_cols.columns, axis=1), df_dummies], axis=1)
pd.set_option('display.max_columns', None)
spaceship_with_dummies.head()
```

**Perform Train Test Split**

```python
target = spaceship_with_dummies["Transported"]

features = spaceship_with_dummies.drop(columns='Transported')
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size = 0.20, random_state=42)
```

```python
normalizer = MinMaxScaler()

normalizer.fit(X_train)
```

```python
X_train_norm = normalizer.transform(X_train)

X_test_norm = normalizer.transform(X_test)
```

```python
X_train_norm = pd.DataFrame(X_train_norm, columns = X_train.columns)
X_train_norm.head()
```

```python
X_train_norm = pd.DataFrame(X_train_norm, columns = X_train.columns)
X_train_norm.head()
```

```python
X_test_norm = pd.DataFrame(X_test_norm, columns = X_test.columns)
X_test_norm.head()
```

- Random Forests

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train_norm, y_train)
```

```python
predictions_RF = model.predict(X_test_norm)
```

- Now let's use the best model we got so far in order to see how it can improve when we fine tune it's hyperparameters.

- Evaluate your model

```python
from sklearn.metrics import accuracy_score

accuracy_RF = accuracy_score(y_test, predictions_RF)
print(f'Accuracy: {accuracy_RF}')
```

**Grid/Random Search**

For this lab we will use Grid Search.

- Define hyperparameters to fine tune.

```python
param_grid = {'n_estimators': [50, 100, 200],
                'max_depth': [None, 10, 20, 30],
                'min_samples_split': [2, 5, 10],
                'min_samples_leaf': [1, 2, 4]}
```

- Run Grid Search

```python
rf = RandomForestClassifier(random_state=42)
```

```python
grid_search = GridSearchCV(estimator=rf, param_grid=param_grid,
                           cv=3, n_jobs=-1, verbose=2)
```

```python
grid_search.fit(X_train_norm, y_train)
```

```python
print(f"Best Parameters: {grid_search.best_params_}")
```

- Evaluate your model

```python
model = RandomForestClassifier(n_estimators=100, max_depth = 20, min_samples_leaf = 4, min_samples_split = 2)
model.fit(X_train_norm, y_train)
```

```python
predictions_RF = model.predict(X_test_norm)
```

```python
accuracy_RF = accuracy_score(y_test, predictions_RF)
print(f'Accuracy: {accuracy_RF}')
```
